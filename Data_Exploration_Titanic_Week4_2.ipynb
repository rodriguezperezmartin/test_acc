```python
# Step1: Import libraries

import pandas as pd
import logging
import seaborn as sns
```

```python
# Step2: Read Dataset

dataset_df = pd.read_csv('./local_data/titanic.csv')
```

```python
#Step3: Data Exploration


print('these are the top 5 rows: ', dataset_df.head())
print('--------------------------------------------')
print('these are last 5 rows: ', dataset_df.tail())
print('--------------------------------------------')
print('these are some random sample rows: ', dataset_df.sample(3))
print('--------------------------------------------')
print('cell done')
```

```python
print('this is the shape of our dataset: ', dataset_df.shape)  #return tupple rows and column cnts
print(' ------------------------------------------ ')
print('this is the info: ', dataset_df.info()) # summary of the DF including labels, data types, non-null values
print(' ------------------------------------------- ')
print('these are descriptive stats for numerical columns, ie. count, mean, std devation: ', dataset_df.describe())
print(' ------------------------------------------- ')
print('these are data types for each column: ', dataset_df.dtypes)
```

```python
##### Exploring Specific Columns #######
print('these are top 5 and bottom 5 records and general facts: ', dataset_df['Survived'].info) #returns top 5 and bottom 5 records, name, num of records and datatype
print('------------------------------------')
print('return the unique values in column: ', dataset_df['Survived'].unique()) #returns the unique values of the column
print('------------------------------------')
print('these are the frequency for each unique value: ', dataset_df['Survived'].value_counts())
print('------------------------------------')
print('counts number of null values in column: ', dataset_df['Survived'].isnull().sum())
```

```python
print('show me all null columns if any: ', dataset_df.isnull().sum())
```

```python
print('frequency for each unique value of Cabin: ', dataset_df['Cabin'].value_counts())
```
